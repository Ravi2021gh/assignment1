```python
# Inheritance is a fundamental concept in object-oriented programming (OOP) that allows a class to inherit properties and behaviors from another class. The class that inherits from another class is called the derived class, subclass, or child class, and the class being inherited from is called the base class, superclass, or parent class. Inheritance promotes code reuse, allows for hierarchical relationships between classes, and enables the creation of more specialized classes based on existing ones.

# There are several types of inheritance in OOP:

# Single Inheritance:
# Single inheritance involves a derived class inheriting from a single base class. It allows the derived class to acquire the attributes and behaviors of the base class. Here's an example:
# python
# Copy code
class Animal:
    def eat(self):
        print("Animal is eating.")

class Dog(Animal):
    def bark(self):
        print("Dog is barking.")

# Create an instance of Dog
my_dog = Dog()
my_dog.eat()  # Inherited from Animal class
my_dog.bark()
# In this example, the Dog class inherits from the Animal class using single inheritance. The Dog class inherits the eat() method from the Animal class and adds its own bark() method.

# Multiple Inheritance:
# Multiple inheritance involves a derived class inheriting from multiple base classes. It allows the derived class to inherit attributes and behaviors from multiple classes. Here's an example:
# python

class Car:
    def drive(self):
        print("Car is driving.")

class Electric:
    def charge(self):
        print("Car is charging.")

class ElectricCar(Car, Electric):
    def self_drive(self):
        print("Electric car is self-driving.")

# Create an instance of ElectricCar
my_electric_car = ElectricCar()
my_electric_car.drive()   # Inherited from Car class
my_electric_car.charge()  # Inherited from Electric class
my_electric_car.self_drive()
# In this example, the ElectricCar class inherits from both the Car and Electric classes using multiple inheritance. It inherits the drive() method from the Car class, the charge() method from the Electric class, and adds its own self_drive() method.

# Multilevel Inheritance:
# Multilevel inheritance involves a derived class inheriting from another derived class. It forms a parent-child-grandchild relationship. Here's an example:
# python

class Vehicle:
    def start(self):
        print("Vehicle is starting.")

class Car(Vehicle):
    def drive(self):
        print("Car is driving.")

class SportsCar(Car):
    def accelerate(self):
        print("Sports car is accelerating.")

# Create an instance of SportsCar
my_sports_car = SportsCar()
my_sports_car.start()     # Inherited from Vehicle class
my_sports_car.drive()     # Inherited from Car class
my_sports_car.accelerate()
# In this example, the Car class inherits from the Vehicle class, and the SportsCar class inherits from the Car class. The SportsCar class inherits the start() method from the Vehicle class, the drive() method from the Car class, and adds its own accelerate() method.

# Hierarchical Inheritance:
# Hierarchical inheritance involves multiple derived classes inheriting from a single base class. It allows multiple classes to inherit from the same base class. Here's an example:
# python

class Animal:
    def eat(self):
        print("Animal is eating.")

class Dog(Animal):
    def bark(self):
        print("Dog is barking.")

class Cat(Animal):
    def meow(self):
        print("Cat is meowing.")

# Create instances of Dog and Cat
my_dog = Dog()
my_dog.eat()   # Inherited from Animal class
my_dog.bark()

my_cat = Cat()
my_cat.eat()   # Inherited from Animal class
my_cat.meow()
   
```
